```python
import requests
from requests import Response

# 下載資料
aqi_url = "https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate%20desc&format=JSON"

res:Response = requests.get(aqi_url)

if res.status_code == 200:
    print("下載成功")
else:
    print("下載失敗")
```

```python
from pydantic import BaseModel, Field, field_validator


class Site_data(BaseModel):
    站點:str = Field(alias="sitename")
    縣市:str = Field(alias="county")
    AQI:int = Field(alias="aqi",default=0)
    空氣品質:str = Field(alias="status")
    污染物:str = Field(alias="pollutant")
    PM25:float = Field(alias="pm2.5")
    so2:float 

    @field_validator("PM25","so2",mode="before")
    @classmethod
    def empty_converter(cls,value:str) -> str:
        if value == "":
            return "0.0"
        else:
            return value

class Aqi_data(BaseModel):
    records:list[Site_data]

aqi_data = Aqi_data.model_validate_json(res.text)

data_list:list[Site_data] = aqi_data.records

print(f"共取出{len(data_list)}筆資料:")
for each_row in data_list:
    print(each_row)
```
